```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type", axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#EarlyStopping ve Dropout sınıfları overfitting problemlerini çözmek için kullanılacak.
#Bu örnekte -görmek açısından- overfitting olacak şekilde model eğitilecektir.
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30, activation="relu")) #units=30 şeklinde de nöron oluşturulabilir.
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=15, activation="relu"))

#Genel geçer bir kural:

#1- İlk katmanın sütun sayısı ile (özellikler ile) aynı olması önerilir,
#2- Aradaki katmanlar ise son katman ile ilk katman arasından bir sayı kadar olması (genelde ortadaki sayı) mantıklı olur.

model.add(Dense(units=1, activation="sigmoid"))
#Son katmana "sigmoid" fonksiyonu koyuldu. Çünkü bu fonksiyon 0-1 arasında değer döndürüyordu.
#Bizim problem tipimiz de sınıflandırma olduğu için dolayısıyla y değerleri 0 ve 1'den oluştuğu için
#son katman bu şekilde oldu.

model.compile(loss="binary_crossentropy", optimizer="adam")
#binary_crossentropy, sınıflandırmada kullanılan bir loss fonksiyonudur.
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1)
```

```python
model.history.history
```

```python
#Görüldüğü üzere aradaki fark çok fazla ve val_loss, normal loss'dan çok daha büyük oldu.
#Yani şu an model overfitting ile eğitim verisini öğrendi.

#Overfitting: (Tekrar)
#Aslında model bizim ana amacımıza ulaşıyor. Yani kendi verilerine benzer bir veri verince çok iyi sonuç veriyor.
#Fakat bu işlemden öte yani 0-1 işlemi yapılacaktan öte elinde bulunduğu verileri yeni verilere uydurmaya çalışıyor.
#Yani kendi içinde tutarlı olmaya çalışırken başka verilerde aşırı tutarsız oluyor.

#Bunu engellemenin bir yolu var:

#Validation_loss yukarı doğru çıkarsa yani artarsa epochs'u durdur, artık train etme, sen bu işin sonuna gelmişsin
#diyebiliyoruz.
#Early Stopping sınıfını bunun için kullanacağız.
```

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=15, activation="relu"))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)

#monitor: Neyi incelememiz gerektiğini belirtir.

#patience: Belli bir epoch'dan sonra modelde herhangi bir iyileştirme olmazsa durdur demektir.
#Çok fazla vermek çok geç durmaya sebep olabilir; çok az vermek çok erken durmaya sebep olabilir.
#20-25 gibi değerler verilebilir, deneme yanılmalıdır, veriye göre değişebilir.

#mode: auto, min ve max şeklinde üçe ayrılır. 
#İnceleyeceğimiz değişken loss olduğu için onu minimumda tut diyeceğiz.
#Başka bir şey takip edildiğinde max veya auto kullanılabilir. (accuracy için max kullanılır)
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])

#Epochs'u yine aynı verdik. patience değerine göre modelde iyileşme olup olmadığının kontrolünü yapacak.
#earlyStopping değişkenimizi callbacks'in içine koymamız gerekiyor.
```

```python
#Görüldüğü üzere model daha fazla iyileşme olmadığına kanaat getirerek eğitimi erkenden durdurdu.
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
#Görüldüğü üzere aradaki mesafe azaldı ve biraz daha iyi.
#Bundan da memnun değilsek Dropout sınıfını kullanacağız.

#rate: 
#Dropout argümanıdır.
#0-1 arasında değer alır. Eğitim esnasında verdiğimiz yüzdeye göre o kadar oranda rastgele katmanları atmaya başlar.
#Bu sayede overfitting'e neden olan katmanlar da atılmış olur, overfitting yüksek oranda engellenir.
```

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))
model.add(Dropout(0.6)) #0.5 ile denenebilir fakat daha üst bir değer verildiğinde uyarı verecektir.

model.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])

#Burada herhangi bir değişiklik yapmaya gerek yok. Gözlemlerimizi yapmaya devam etmemiz gerekiyor.
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```

```python
#Görüldüğü üzere aradaki mesafe çok azaldı ve daha iyi bir sonuç verdi.
#Not: Dropout kullanıldığı için loss'un zikzaklı olması çok normal.

#Daha da optimize edilebilir fakat şu an yeterli seviyededir.
```

```python
tahminlerimiz = model.predict_classes(x_test) #Sınıflandırma örneğinde "predict_classes" kullanılarak tahmin edilir.
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#confusion matrix: Sınıflandırmamızın ne kadar hatalı sonuçlar verdiğini gösteren bir matristir.
#classification_report: Sınıflandırmamızın genel ölçülerini gösteren bir matristir.
```

```python
print(classification_report(y_test, tahminlerimiz))
```

```python
#Bu matriste önemli olan kısım precision kısmıdır.
#0 ve 1'leri ne kadar oranda tahmin ettiğini gösterir.
#Burada %90 gibi çok iyi bir oranda tahmin ettiğini gösteriyor.
```

```python
print(confusion_matrix(y_test, tahminlerimiz))
```
