```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#Bu komut, yazdığımız grafikleri burada kolayca gösterilmeye olanak tanır.
#Aslında bunu yazmadan "plt.show()" diyerek de çizdirilebilir fakat bunu azaltmak için yazdık.
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)

#Numpy dizileri ile çalışılması tavsiye edilir. Çünkü Numpy'ın hazır komutlarını kullanmak isteyebiliriz.
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yas") #x ekseninin ismi
plt.ylabel("Kilo") #y ekseninin ismi
plt.title("Kilonun Yaşa Göre Değişim Grafiği") #Grafiğin başlığı
#plt.show()

#Çizdirmek için genel kullanım: plt.plot(x ekseni, y ekseni, çizginin rengi ve grafik türü, ...)
#g burada green demek. İstenilen rengin baş harfi ya da tam ismi yazılabilir (r, b, y veya red, blue, yellow vs).
```

```python
numpyDizisi1 = np.linspace(0,10,20)

#Hatırlatma: (0,10) arasında "eşit mesafede" 20 elemanlı bir matris oluştur demektir. (Sütun sayısı her zaman 5'tir.)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
#plt.plot(numpyDizisi1,numpyDizisi2,"g*-") --> yeşil renkte yıldızlı çizgi grafiği oluştur anlamına gelir.
#plt.plot(numpyDizisi1,numpyDizisi2,"b*-") --> mavi renkte yıldızlı çizgi grafiği oluştur anlamına gelir.
#gibi grafikler çizdirilebilir. (renk+grafik türü) şeklinde.

#Bilgi: Konveks grafik oluştu. Verilerin yeri değiştiğinde grafik konkav olur.
```

```python
plt.subplot(1,2,1) #1 satır 2 sütun olacak şu an 1. grafiği çiziyorum.
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) #1 satır 2 sütun olacak şu an 2. grafiği çiziyorum.
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
#plt.subplot veya plt.plot diyip işlem yapmak yerine kendi "figure"ümüzü de oluşturabiliriz.
#Yani biz "plt" olarak kullandığımız matplotlib kütüphanesinden bir obje oluşturup 
#yapmasını istediğimiz özel işlemler de yaptırabiliriz.
```

```python
benimFigure = plt.figure()
#Figure dediğimiz şey aslında grafik çizdirdiğimiz dikdörtgensel alan(kanvas).
#Bu komutun yapılacak işlemlerin olduğu hücrede olması tavsiye edilir.

figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
#Boyutlar girildi. [X Noktası, Y Noktası, Genişlik, Yükseklik]
#İçe içe grafik oluştururken x ve y noktaları anlam kazanacak.

figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9]) #[X Noktası, Y Noktası, Genişlik, Yükseklik]
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
#subplots() metodunun başka bir kullanımı:
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2)
benimEksenler.plot(numpyDizisi1,numpyDizisi2,"b")

#Normalde plt.subplots() bir tuple döndürdüğü için "(benimFigure, benimEksen)" şeklinde yazdık.
#Yani normalde plt.figure() yazdığımızda yalnız figure döndürürken subplots() yazdığımızda her ikisini de döndürüyor.
#benimEksenler artık tek bir eksen olmadığı ve bir dizi olduğu için hata verdi ve çizdirmedi.
#Grafiği çizdirebilmemiz için for döngüsü kullanmamız gerekiyor.
```

```python
type(benimEksenler)
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")

plt.tight_layout() #İki grafik arasındaki mesafeyi arttırır.
#benimFigure
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 2, label = "numpyDizisi ** 2") #Çizginin etiketleri belirtildi.
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc=1) #Etiketleri göstermek için bu komut kullanılmalıdır.
#İstendiğinde etiketin nerede olacağını loc="sayı" yazılarak belirtilebilir. ((0,10) aralığında)
```

```python
#Figürün boyutunu "add_axes" şeklinde yazabildiğimiz gibi figure() metodunun içerisine de figsize şeklinde ekleyebiliriz.
```

```python
yeniFigur = plt.figure(figsize=(3,3), dpi=100) #dpi argümanı ile de grafiğin kalitesini belirleyebiliriz.

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
#Aslında bunu yazmaya gerek yok. Çünkü boyutlarını figüre içerisine girdik.

yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc=1)
```

```python
yeniFigur.savefig("benimfigur.png", dpi=200)

#Grafiği kaydetmek için savefig() metodu kullanılır.
#dpi argümanını dilersek dosyayı kaydederken de yazabiliriz.
```
