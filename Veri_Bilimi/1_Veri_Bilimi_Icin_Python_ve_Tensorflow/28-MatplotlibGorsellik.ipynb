```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2, color="#4A7BFE", alpha=0.3) #Aynı hücrede yazmamız gerektiğini unutmamalıyız.
benimEksen.plot(numpyDizisi2, numpyDizisi1, color="#CD6135")

#İstediğimiz rengi eklemek için Google'a "rgb hex code" yazıp o renk kodunu color="" tırnak içerisine ekleyebiliriz.
#alpha ile rengin saydamlığını belirleyebiliriz.
```

```python
(yeniFigur, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color="blue", linewidth = 8.0) #linewidth: Çizgi kalınlığı ayarlama.
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, color="yellow", linewidth = 6.0)

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color="#CD6135", linestyle = "--") #linesytle: Çizgi stilini ayarlama.
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color="#343A49", linestyle = ":")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 6, color="#A444FF", linestyle = "-.")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7, color="black", linestyle = "--", marker="o", markersize=8,markerfacecolor="red")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8, color="black", linestyle = "--", marker="+", markersize=4)

#marker: (x,y) noktasının işaretini ayarlama.
#markersize: (x,y) noktasının boyutunu ayarlama.
#markerfacecolor: (x,y) noktasının rengini ayarlama (eğer içi doldurulabiliyorsa çalışır).
```
